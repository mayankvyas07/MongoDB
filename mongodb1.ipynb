```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios is it preferred to use MongoDB over SQL databases?

MongoDB is a popular open-source NoSQL database management system that stores data in a flexible, JSON-like format known as BSON (Binary JSON). It is designed for handling large volumes of unstructured or semi-structured data and offers high performance, scalability, and flexibility.

Non-relational databases (NoSQL) are databases that do not rely on the traditional tabular relational model used by SQL databases. Instead, they use various data models like key-value, document, column-family, or graph, which are better suited for specific types of data and applications. NoSQL databases provide flexibility, horizontal scalability, and can handle large amounts of diverse data.

Scenarios where MongoDB is preferred over SQL databases include:

When dealing with large amounts of unstructured or semi-structured data.
Applications with rapidly changing schemas or requirements.
Need for high-speed data retrieval and real-time analytics.
Projects requiring horizontal scalability and distribution.
Agile development environments where frequent updates are common.
Q2. State and Explain the features of MongoDB.

MongoDB offers several features, including:

Flexible Schema: MongoDB allows documents within a collection to have varying structures, making it suitable for evolving applications.
NoSQL Nature: Data is stored in BSON format (Binary JSON), which supports nested arrays and documents.
High Performance: MongoDB provides fast read and write operations due to its memory-mapped storage engine.
Horizontal Scalability: Data can be distributed across multiple servers to handle high loads.
Indexing: Supports various types of indexing for efficient data retrieval.
Aggregation Framework: Allows complex data aggregation and transformation operations.
Replication and Fault Tolerance: Offers automatic replication and failover to ensure data availability.
JSON-like Documents: Documents are stored in a human-readable JSON-like format (BSON).
Rich Query Language: Provides powerful query capabilities using a query language similar to SQL.
Geospatial Indexing: Supports geospatial data and queries.
Ad Hoc Queries: Developers can query the database without the need to define the schema in advance.
Q3. Write code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

python
Copy code
import pymongo

# Connect to MongoDB server (default connection is to localhost)
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create/connect to a database
mydb = client["mydatabase"]

# Create/connect to a collection within the database
mycollection = mydb["mycollection"]
Q4. Using the database and the collection created in question number 3, write code to insert one record and insert many records. Use the find() and find_one() methods to print the inserted record.

python
Copy code
# Insert one record
record = {"name": "Alice", "age": 30}
inserted_record = mycollection.insert_one(record)
print("Inserted ID:", inserted_record.inserted_id)

# Insert many records
records = [
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 28}
]
inserted_records = mycollection.insert_many(records)
print("Inserted IDs:", inserted_records.inserted_ids)

# Print the inserted record using find_one()
found_record = mycollection.find_one({"name": "Alice"})
print("Found Record:", found_record)
Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

The find() method in MongoDB is used to query the database for documents that match specific criteria. It returns a cursor that can be iterated to retrieve matching documents.

python
Copy code
# Query and print all documents in the collection
cursor = mycollection.find({})
for document in cursor:
    print(document)
Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

The sort() method in MongoDB is used to specify sorting order for query results. It takes a dictionary where keys are field names and values are either 1 (ascending) or -1 (descending).

python
Copy code
# Sort documents by age in descending order
sorted_cursor = mycollection.find({}).sort("age", -1)
for document in sorted_cursor:
    print(document)
Q7. Explain why delete_one(), delete_many(), and drop() are used.

delete_one(filter): This method deletes a single document from the collection that matches the given filter criteria.
delete_many(filter): This method deletes all documents from the collection that match the given filter criteria.
drop(): This method is used to delete an entire collection, including all its documents and indexes.
```
